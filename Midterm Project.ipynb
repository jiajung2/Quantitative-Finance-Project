```python
# Import the pandas library as pd
import pandas as pd

# Import the plotting library matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import statsmodels.api as sm

plt.style.use('classic')
%matplotlib inline
```

# GDP

```python
# Download GDP data
GDP= pd.read_csv('https://fred.stlouisfed.org/data/GDP.txt',skiprows=18,sep='\s+',index_col=0,parse_dates = True)
# set cpi equal to 'VALUE' column of GDP
GDP_data= GDP['VALUE']

# First 5 rows of GDP
GDP.head()
```

```python
# Create figure and axis
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# Plot
ax.plot(GDP_data[-21:],'-',lw=3,label='Gross Domestic Income')

# Add title and y-axis label
ax.set_title('Gross Domestic Income')
ax.set_ylabel('Billions of Dollars')

# Auto-rotate the x-axis labels
fig.autofmt_xdate()

ax.grid()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Save the figure as a png file. bbox_inches = 'tight' reduces whitespace around the figure
fig.savefig('GDP.png',dpi = 120,bbox_inches='tight')
```

```python
# Compute GDP Growth rate
GDP_growth = GDP_data/GDP_data.shift(1)-1

# Create figure and axis
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# Plot
ax.plot(100*4*GDP_growth[-21:],'-',lw=3,label='GDP Growth Rate')

# Add title and y-axis label
ax.set_title('GDP Growth Rate')
ax.set_ylabel('Percent (Annualized)')

# Auto-rotate the x-axis labels
fig.autofmt_xdate()

ax.grid()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Save the figure as a png file. bbox_inches = 'tight' reduces whitespace around the figure
fig.savefig('GDP Growth Rate.png',dpi = 120,bbox_inches='tight')
```

```python
# Compute and print average GDP Growth rates
print('GDP Growth Rate over previous year:     ',round(100*(GDP_data.iloc[-1]/GDP_data.iloc[-1-1*4]-1),5),'%')
print('GDP Growth Rate over previous 2 years:  ',round(100*((GDP_data.iloc[-1]/GDP_data.iloc[-1-2*4])**(1/2)-1),5),'%')
print('GDP Growth Rate over previous 10 years: ',round(100*((GDP_data.iloc[-1]/GDP_data.iloc[-1-10*4])**(1/10)-1),5),'%')
print('GDP Growth Rate over previous 50 years: ',round(100*((GDP_data.iloc[-1]/GDP_data.iloc[-1-50*4])**(1/50)-1),5),'%')
print('GDP Growth Rate over previous all years:', round(100*((GDP_data.iloc[-1]/GDP_data.iloc[0])**(1/(len(GDP_data)-1)*12)-1),5),'%')
```

Real gross domestic product (GDP) is GDP given in constant prices and refers to the volume level of GDP

Because GDP is one of the most important metrics for evaluating the economic activity, stability, and growth of goods and services in an economy, it is usually reviewed from two angles: nominal and real. Nominal GDP is a macroeconomic assessment of the value of goods and services using current prices in its measure; it's also referred to as the current dollar GDP.

Real GDP takes into consideration adjustments for changes in inflation. This means that if inflation is positive, real GDP will be lower than nominal, and vice versa. Without a real GDP adjustment, positive inflation greatly inflates GDP in nominal terms.

```python
c=GDP_business.rename(columns = {"VALUE":"GDP business Sector"})
d=GDP_household.rename(columns = {"VALUE":"GDP Household Sector"})
e=GDP_Government.rename(columns={"VALUE":"GDP Government Sector"})
df2=pd.concat([c, d,e], axis=1)
```

```python
df2.head()
```

```python
# Create a variable 'X' to be the independent variable of the OLS model. Do not add a constant to X
X = df2['GDP business Sector']

# Create a variable 'Y' to be the dependent variable of the OLS model.
Y = df2['GDP Household Sector']

# Create a variable called 'model' that initializes the OLS model
model = sm.OLS(Y,X)

# Create a variable called 'results' that stores the results of the estimated OLS model
results = model.fit()

# Print output of summary() method of results
print(results.summary())
```
